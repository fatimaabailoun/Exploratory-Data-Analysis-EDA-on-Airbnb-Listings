# Cleaning Data

Remove All UnNeeded Columns

```python
data = df.drop(['listing_url','scrape_id', 'last_scraped','source','picture_url','host_url','name','host_location','host_thumbnail_url','host_picture_url','host_listings_count','host_neighbourhood','property_type','bathrooms','bedrooms','calculated_host_listings_count_shared_rooms','calculated_host_listings_count_private_rooms','calculated_host_listings_count_entire_homes','calculated_host_listings_count','license','review_scores_value','review_scores_location','review_scores_communication','review_scores_checkin','review_scores_cleanliness','review_scores_accuracy','last_review','first_review','number_of_reviews_l30d','number_of_reviews_ltm','number_of_reviews','calendar_last_scraped','availability_365','availability_90','availability_60','availability_30','has_availability','calendar_updated','maximum_nights_avg_ntm','minimum_nights_avg_ntm','maximum_maximum_nights','minimum_maximum_nights','maximum_minimum_nights','minimum_minimum_nights','neighbourhood_group_cleansed','neighbourhood'] , axis=1 )
data.head()
```

Handle Missing Values

```python
data.info() #check null values
```

```python
data.isnull().sum() #another way ti check the exact number of present null values 
```

```python
#replacing null values in some columns by zero or 'non' value
data['neighborhood_overview']=data['neighborhood_overview'].fillna('Non')
data['host_about']=data['host_about'].fillna('Non')
data['host_response_rate']=data['host_response_rate'].fillna(0)
data['host_acceptance_rate']=data['host_acceptance_rate'].fillna(0)
data['review_scores_rating']=data['review_scores_rating'].fillna(0)
data['reviews_per_month']=data['reviews_per_month'].fillna(0)
```

```python
#deleting rows that contain null values (in some columns we ca't replace null values since we need a real value)
data1 = data.dropna(axis=0)
```

```python
data1.isnull().sum() #Re-check null values
```

Check Duplicated values

```python
data1.duplicated() #no duplicated values
```

Rename Columns

```python
data1.rename(columns = {'description':'Place Description'}, inplace = True)
data1.rename(columns = {'neighborhood_overview':'Neighborhood Description'}, inplace = True)
data1.rename(columns = {'bathrooms_text':'bathroom'}, inplace = True)
data1.rename(columns = {'neighbourhood_cleansed':'neighbourhood'}, inplace = True)
data1.head()
```

convert price column from string to int to be able to apply statistics

```python
data1['price'] = pd.to_numeric(data1['price'], errors='coerce') #values are converted from string to float
#errors='coerce' parameter tells pandas to replace any non-numeric values with NaN
data1['price'] = data1['price'].astype('Int64') #values are converted from float to int
```
