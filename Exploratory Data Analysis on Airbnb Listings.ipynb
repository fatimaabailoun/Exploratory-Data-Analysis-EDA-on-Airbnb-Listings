# Cleaning Data

Remove All UnNeeded Columns

```python
data = df.drop(['listing_url','scrape_id', 'last_scraped','source','picture_url','host_url','name','host_location','host_thumbnail_url','host_picture_url','host_listings_count','host_neighbourhood','property_type','bathrooms','bedrooms','calculated_host_listings_count_shared_rooms','calculated_host_listings_count_private_rooms','calculated_host_listings_count_entire_homes','calculated_host_listings_count','license','review_scores_value','review_scores_location','review_scores_communication','review_scores_checkin','review_scores_cleanliness','review_scores_accuracy','last_review','first_review','number_of_reviews_l30d','number_of_reviews_ltm','number_of_reviews','calendar_last_scraped','availability_365','availability_90','availability_60','availability_30','has_availability','calendar_updated','maximum_nights_avg_ntm','minimum_nights_avg_ntm','maximum_maximum_nights','minimum_maximum_nights','maximum_minimum_nights','minimum_minimum_nights','neighbourhood_group_cleansed'] , axis=1 )
data.head()
```

Handle Missing Values

```python
data.info() #check null values
```

```python
data.isnull().sum() #another way to check the exact number of present null values 
```

```python
#replacing null values in some columns by zero or 'non' value
data['neighborhood_overview']=data['neighborhood_overview'].fillna('Non')
data['host_about']=data['host_about'].fillna('Non')
data['host_response_rate']=data['host_response_rate'].fillna(0)
data['host_acceptance_rate']=data['host_acceptance_rate'].fillna(0)
data['review_scores_rating']=data['review_scores_rating'].fillna(0)
data['reviews_per_month']=data['reviews_per_month'].fillna(0)
```

```python
#deleting rows that contain null values (in some columns we ca't replace null values since we need a real value)
data = data.dropna(axis=0)
```

```python
data.isnull().sum() #Re-check null values
```

Check Duplicated values

```python
data.duplicated() #no duplicated values
```

Rename Columns

```python
data.rename(columns = {'description':'Place Description'}, inplace = True)
data.rename(columns = {'neighborhood_overview':'Neighborhood Description'}, inplace = True)
data.rename(columns = {'bathrooms_text':'bathroom'}, inplace = True)
data.rename(columns = {'neighbourhood_cleansed':'neiborhood'}, inplace = True)
data.head()
```

convert price column from string to int to be able to apply statistics

```python
# Remove non-numeric characters and convert string to int data in column price 
data['price'] = data['price'].astype(str).str.replace('[$,]', '', regex=True).str.split('.').str[0].astype(int)
```

```python
List_Price_mean = data[['room_type','price']].groupby('room_type').mean()
List_Price_mean #calculate the mean of price with relation to Listings
```

```python
room_by_neighborhood = data['neiborhood'].value_counts()
room_by_neighborhood #counting the nb of rooms per neiborhood
```

```python
# Correlation Analysis between Price and nb of bed per room
accommodates_price_correlation = data[['price', 'accommodates']].corr()
accommodates_price_correlation
```

This value (0.33718) suggests a positive relationship between number of accommodates per listing and the price of it. However, since the value is closer to 0 than 1, it indicates a relatively weak positive correlation. This means that there is some tendency for the two entities to move in the same direction, but it's not a strong or highly reliable relationship. Other factors may also influence their behavior.
